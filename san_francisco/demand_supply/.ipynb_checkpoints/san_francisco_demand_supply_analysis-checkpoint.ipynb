### Datasets A & B: Assumptions

**Assumption 1**<br/>Outliers should be removed only for the rides dataset, and should be based on ride startedLocation longitudes and/or latitudes being out of bounds. Outliers do not need to be removed from the sessions dataset.<br/><br/>
**Assumption 2**<br/>There is some relationship between userId, startedAt, and endedAt fields that can help join the datasets and create a new boolean variable to determine whether or not a session led to a ride. One hypothesis we can test is that a ride's time period must take place within a session's time period. If we can verify this hypothesis, then we can join the two datasets together.

#### Exploring Assumption 2

```python
# Testing a single userId chosen at random
userId_selected = '0BqhtgFXvL7LhBnuv/wKUyPLb4DUEcy1kPvLITrZOtI='
df_session_sample = df_sessions_final[df_sessions_final['userId']==userId_selected]
df_ride_sample = df_rides_final[df_rides_final['userId']==userId_selected]
```

```python
display(df_session_sample)
```

```python
display(df_ride_sample)
```

From the sample data we can see that for the sample user...<br/>
* On 2019-05-03, had a 100% success rate of a session leading to a ride<br/>
* On 2019-05-04, had a 50% success rate of a session leading to ride, where the user first searched for ~1 min then abandoned the session, but then started a new session ~40 minutes later and then was able to ride<br/>
* The user can have a session open for many minutes before starting a ride
* Our hypothesis for Assumption 2 appears to be true, we can continue to test the join

### Dataset C: Sessions + Rides + Join Based on Assumption 2

```python
# Merge the two datasets with a userId join
df_merged = df_sessions_final.merge(df_rides_final,on='userId',how='left')
display(df_merged.head(2))

# Notes
# df_sessions_final startedAt is renamed to startedAt_x and endedAt is renamed to endedAt_x
# df_rides_final startedAt is renamed to startedAt_y and endedAt is renamed to endedAt_y
```

```python
# Filter data where Assumption 2 conditions met
df_joined_as2 = df_merged[(df_merged.startedAt_x <= df_merged.startedAt_y) & (df_merged.endedAt_x >= df_merged.endedAt_y)]

# Renaming Columns
df_joined_as2.rename(columns={'startedAt_x': 'startedAt', 'endedAt_x': 'endedAt', 'startedAt_y': 'startedAtRide', 'endedAt_y': 'endedAtRide'}, inplace=True, errors='ignore')

# Checking record count
print('Assumpted 2, joined: ','\t',len(df_joined_as2))
print('Compare pre-joined: ','\t',count_rides)
```

It looks like from the comparison of records pre- and post-join that there are rides that exist without following Assumption 2.

One preliminary strategy we can take is to keep Assumption 2 as it is, we can create more comprehensive logic to connect sessions and rides with subsequent iterations.

```python
display(df_joined_as2.head(2))
```

```python
# Strategy to merge all the data together to obtain complete joined dataset
df_c = df_joined_as2 + df_sessions_final# + df_rides_final

df_c = pd.concat([df_joined_as2,df_sessions_final,df_rides_final], axis=0, join='outer', sort=False, copy=True)

display(df_c.head(2))
```

```python
# Remove unnecessary columns
df_c.drop(columns=['geometry_x','geometry_y','geometry'], inplace=True)

df_c.drop_duplicates(subset=['userId','startedAt','endedAt','sessionStartLat','sessionStartLong',
                             'startedAtRide','endedAtRide','startedLocationLat','startedLocationLong',
                             'endedLocationLat','endedLocationLong'], inplace=True)

display(df_c.head(5))
```

```python
df_c.describe()
```

```python
print('Count Rides:','\t','\t',count_rides)
print('Count Sessions:','\t',count_sessions)
```

We can see that our new df_c combined dataset is now a combination of the original datasets (without outliers) and the Assumption 2 joined dataset. Let's see what this looks like for the sample userId.

```python
# Testing a single userId chosen at random
df_c_sample = df_c[df_c['userId']==userId_selected]

display(df_c_sample)
```

```python
# For this exercise it will be useful to have the three separate categories of data for visualization purposes

# Filling NaN's with 0's
df_c.fillna(value=0,inplace=True)

# Creating a new column to identify the category of data

# Function to categorize types
def type_identify(row):
    if (int(row['sessionStartLat']) != 0) & (int(row['startedLocationLat']) != 0):
        val = 'AS2 Joined'
    elif (int(row['sessionStartLat']) != 0) & (int(row['startedLocationLat']) == 0):
        val = 'Sessions Only'
    elif (int(row['sessionStartLat']) == 0) & (int(row['startedLocationLat']) != 0):
        val = 'Rides Only'
    else:
        val = 'N/A'
    return val

df_c['type'] = df_c.apply(type_identify, axis=1)
```

```python
# Testing a single userId chosen at random
df_c_sample = df_c[df_c['userId']==userId_selected]

display(df_c_sample)
```

### Dataset D: Sessions + Rides + Join Based on Assumption 2 + Neighborhoods

```python
df_d = df_c.copy()
```

```python
# Source: http://mattmurray.net/unlocking-the-power-of-geospatial-data-with-geopandas/
# Source: https://medium.com/dataexplorations/working-with-open-data-shape-files-using-geopandas-how-to-match-up-your-data-with-the-areas-9377471e49f2

# Ingest the .geojson file(s)
df_nbh = gpd.read_file(folder_maps + 'san-francisco.geojson')

display(df_nbh.head(3))
```

```python
# Session

# Setting points
geometry_d_session = [Point(xy) for xy in zip(df_d['sessionStartLong'], df_d['sessionStartLat'])]

# Specifying CRS
crs = "EPSG:4326"

# Creating the GeoDataFrame
geo_locs_d_session = gpd.GeoDataFrame(df_d, crs=crs, geometry=geometry_d_session)

# Sampling
display(geo_locs_d_session.head(3))
```

```python
# Note: Make sure you have run pip3 install rtree && brew install spatialindex
# Using a spatial join to create new column neighborhood

# For Session
comb_d_session = gpd.sjoin(geo_locs_d_session, df_nbh, how='left')
comb_d_session.rename(columns={'name': 'session_neighborhood'}, inplace=True)
comb_d_session.drop(columns=['cartodb_id','created_at','updated_at','index_right','geometry'], inplace=True)
display(comb_d_session.head(3))
```

```python
# Check
print(len(comb_d_session))
comb_d_session.groupby(['type']).count()
```

```python
# For Ride Start

# Setting points
geometry_d_ride_start = [Point(xy) for xy in zip(comb_d_session['startedLocationLong'], comb_d_session['startedLocationLat'])]

# Specifying CRS
crs = "EPSG:4326"

# Creating the GeoDataFrame
geo_locs_d_ride_start = gpd.GeoDataFrame(comb_d_session, crs=crs, geometry=geometry_d_ride_start)

# Sampling
display(geo_locs_d_ride_start.head(3))
```

```python
# For Ride Start

comb_d_session_ride_start = gpd.sjoin(geo_locs_d_ride_start, df_nbh, how='left')
comb_d_session_ride_start.rename(columns={'name': 'ride_start_neighborhood'}, inplace=True)
comb_d_session_ride_start.drop(columns=['cartodb_id','created_at','updated_at','index_right','geometry'], inplace=True)
display(comb_d_session_ride_start.head(3))
```

```python
# Check
print(len(comb_d_session_ride_start))
comb_d_session_ride_start.groupby(['type']).count()
```

```python
# For Ride End

# Setting points
geometry_d_ride_end = [Point(xy) for xy in zip(comb_d_session_ride_start['endedLocationLong'], comb_d_session_ride_start['endedLocationLat'])]

# Specifying CRS
crs = "EPSG:4326"

# Creating the GeoDataFrame
geo_locs_d_ride_end = gpd.GeoDataFrame(comb_d_session_ride_start, crs=crs, geometry=geometry_d_ride_end)

# Sampling
display(geo_locs_d_ride_end.head(3))
```

```python
# For Ride End

comb_d_session_ride_start_end = gpd.sjoin(geo_locs_d_ride_end, df_nbh, how='left')
comb_d_session_ride_start_end.rename(columns={'name': 'ride_end_neighborhood'}, inplace=True)
comb_d_session_ride_start_end.drop(columns=['cartodb_id','created_at','updated_at','index_right','geometry'], inplace=True)
display(comb_d_session_ride_start_end.head(3))
```

```python
# Check
print(len(comb_d_session_ride_start))
comb_d_session_ride_start.groupby(['type']).count()
```
